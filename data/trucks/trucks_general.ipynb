```python
import json
import re
import pandas as pd
import urllib
from urllib.request import Request
from bs4 import BeautifulSoup
```

```python
url = "https://www.otomoto.pl/ciezarowe?page="

pages = []
for page_number in range(1, 408):
    print(f'Pobieranie strony {page_number}')
    with urllib.request.urlopen(f'{url}{page_number}') as resp:
        processed_page = BeautifulSoup(resp.read().decode('utf-8'), "html.parser")
        pages.append(processed_page)
```

```python
print(pages[0].prettify())
```

```python
general_truck_data = []
for page in pages:
     general_truck_data += page.find_all('article', attrs={'class': lambda c: c.startswith('ooa'), 'data-testid': 'listing-ad'})

print(len(general_truck_data))  
print(general_truck_data[0])
```

```python
list_of_voivodeships = []
for page in pages:
    spans = page.find_all('span', class_='ooa-fzu03x')
    for span in spans:
        voivodeship = span.contents[4].strip()
        list_of_voivodeships.append(voivodeship)

#tutaj mamy listę wszystkich ogłoszeń, czyli mozemy sobie z tego
#wyciągnać liste wszystkich ogloszen wedlug wojewodztwa


#tutaj robimy unikalne
unique_voivodeships = list(set(list_of_voivodeships))
print(unique_voivodeships)
print(len(unique_voivodeships))

print("Unikalne wojewodztwa lacznie")
#print(list_of_voivodeships)
print(len(list_of_voivodeships))
```

```python
list_of_cities = []
for page in pages:
    spans = page.find_all('span', class_='ooa-fzu03x')
    for span in spans:
        city = span.contents[0].strip()
        list_of_cities.append(city)

#tutaj mamy listę wszystkich ogłoszeń, czyli mozemy sobie z tego
#wyciągnać liste wszystkich ogloszen wedlug miejscowosci

#tutaj robimy unikalne
unique_cities = list(set(list_of_cities))
print(unique_cities)
print("Unikalne miasta łącznie")
print(len(unique_cities))

#print(list_of_cities)
print("Wszystkie miasta łącznie")
print(len(list_of_cities))
```

```python
truck_names = []

motorbikes_elements = pages[0].find_all('em')
for motorbike_element in motorbikes_elements:
    motorbike_name = motorbike_element.get_text(strip=True)
    truck_names.append(motorbike_name)

unique_truck_names = list(set(truck_names))
unique_truck_names = [car for car in  list(set(truck_names)) if car not in list_of_voivodeships]
print(unique_truck_names)
```

```python
truck_info = []

for i in range(1,19264):
    #print(general_delivery_car_data[i])
    
    name_element = general_truck_data[i].find('h2', class_='evg565y6')
    
    if name_element is None:
        continue  # Skip this iteration if name element is not found
    
    name = name_element.text.strip()
    model = ''
    
    for unique_name in unique_truck_names:
        if unique_name in name:
            model = name.replace(unique_name, '').strip()
            name = unique_name
            break
    # Extracting price
    price_element = general_truck_data[i].find('div', class_='evg565y12')
    
    if price_element is None:
        continue  # Skip this iteration if price element is not found
    
    price = price_element.text.strip()
    
    # Extracting city and voivodeship
    location_element = general_truck_data[i].find('span', class_='ooa-1ammnlq')
    
    if location_element is None:
        continue  # Skip this iteration if location element is not found
    
    city_element = location_element.find('span', class_='ooa-fzu03x')
    
    if city_element is None:
        continue  # Skip this iteration if city element is not found
    
    city = city_element.text.strip().split('(')[0].strip()
    voivodeship = city_element.text.strip().split('(')[1].strip().strip(')')
    
    # Extracting production year, mileage, engine capacity, and engine type
    details = general_truck_data[i].find_all('li', class_='ooa-1k7nwcr')
    
    if len(details) < 4:
        continue  # Skip this iteration if details are incomplete
    
    production_year = details[0].text.strip()
    mileage = details[1].text.strip()
    engine_capacity = details[2].text.strip()
    engine_type = details[3].text.strip()
    
    # Creating a dictionary for the current car offer
    car_offer = {
        'Name': name,
        'Model': model,  # do zrobienia z Name
        'Price': price,
        'City': city,
        'Voivodeship': voivodeship,
        'Production Year': production_year,
        'Mileage': mileage,
        'Engine Type': engine_type,
        'Engine Capacity': engine_capacity
    }
    
    # Adding the car offer to the delivery_car_data array
    truck_info.append(car_offer)
print(len(truck_info))
```

```python
frame = pd.DataFrame.from_dict(truck_info)
frame
```

```python
import csv
filename = 'truck_info.csv'

with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = [
        'Name',
        'Model',
        'Price',
        'City',
        'Voivodeship',
        'Production Year',
        'Mileage',
        'Engine Type',
        'Engine Capacity'
    ]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()  # Write the header row
    
    for truck_offer in truck_info:
        writer.writerow(truck_offer)
    
print(f"Data saved to {filename} successfully.")
```
